```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(["name","body_type","national_team","national_rating","national_team_position","national_jersey_number","release_clause_euro"],axis=1, inplace = True)
```

```python
df2 = df.dropna(subset=["wage_euro","value_euro"])
```

```python
df2.isnull().sum()
```

### Total Number of Players for Each Country

```python
#Her √ºlke i√ßin toplam futbolcu sayƒ±sƒ±
df2.groupby("nationality").count()[["overall_rating"]].sort_values('overall_rating', axis=0,ascending=False)
```

### Most Valuable Player of Every Country

```python
#her √ºlke i√ßin en deƒüerli futbolcu
df_filtered = df[['nationality', 'full_name', 'value_euro']].dropna()
top_players_for_each_country = df_filtered.loc[df_filtered.groupby('nationality')['value_euro'].idxmax()].sort_values("value_euro", ascending=False).reset_index()
top_players_for_each_country
```

### Most Valueable Players by Age

```python
#her ya≈ü i√ßin en deƒüerli futbolcular
top_player_for_each_age = df2.loc[df2.groupby("age")["value_euro"].idxmax().round().astype(int),["full_name","age","overall_rating","value_euro"]].sort_values("age",ascending = True)
top_player_for_each_age
```

### Average Market Value by Age

```python
#her ya≈ü i√ßin ortalama deƒüerler
avg_values_for_each_age = df2.groupby("age")["value_euro"].mean().round().astype(int).reset_index()
avg_values_for_each_age
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="age", y="value_euro", data = avg_values_for_each_age, hue="value_euro", palette="viridis")

plt.title("Average Market Value by Age (‚Ç¨)", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average Market Value (‚Ç¨)", fontsize=12)
plt.xticks(rotation=45) 
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', '.')))

plt.show()
```

### Average Wages by Age

```python
avg_wage_for_each_age = df2.groupby("age")["wage_euro"].mean().round().astype(int).reset_index()
avg_wage_for_each_age
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="age",y="wage_euro",data = avg_wage_for_each_age, hue="wage_euro", palette = "viridis")

plt.title("Average Wages by Age")
plt.xlabel("Age")
plt.ylabel("Average Wages")
```

### Correlation Between Wages and Overall Ratings

```python
filtered_wages_overall = df2[["wage_euro","overall_rating"]].round().astype(int)
correlation = filtered_wages_overall.corr().iloc[0,1]

correlation

plt.figure(figsize=(10, 6))
sns.scatterplot(x = filtered_wages_overall['overall_rating'], y = filtered_wages_overall['wage_euro'])
sns.regplot(x = filtered_wages_overall['overall_rating'], y = filtered_wages_overall['wage_euro'], scatter_kws={'alpha':0.5}, line_kws={"color": "green"})
plt.xlabel("Overall Rating")
plt.ylabel("Wage (Euro)")
plt.title(f"Overall Rating vs Wage (Correlation: {correlation:.2f})")
plt.show()
```

### Data Preparation Rating and Skills Analysis

```python
df2.columns
```

```python
df3 = df2[["full_name","overall_rating","positions",'weak_foot(1-5)', 'skill_moves(1-5)', 'crossing', 'finishing',
       'heading_accuracy', 'volleys', 'dribbling', 'curve',
       'freekick_accuracy', 'ball_control', 'acceleration',
       'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power',
       'jumping', 'stamina', 'strength', 'long_shots', 'aggression',
       'interceptions', 'positioning', 'vision', 'penalties', 'composure',
       'marking', 'standing_tackle', 'sliding_tackle']].copy()

avg_passing = (df2["short_passing"] + df2["long_passing"]) / 2
df3["average_passing"] = avg_passing

#Getting main position
df3["positions"] = df3["positions"].str.split(',').str[0]

df3
```

```python
df3.groupby("positions").size().sort_values(ascending=False)
```

```python
df3.info()
```

### Correlation Between Overall Rating and Skills

```python
# Sayƒ±sal s√ºtunlarƒ± se√ßiyoruz (full_name ve positions hari√ß)
numeric_cols = df3.select_dtypes(include=['int64', 'float64']).columns
numeric_df = df3[numeric_cols]

# Overall rating ile diƒüer √∂zelliklerin korelasyon hesaplamasƒ±
correlations = numeric_df.corr()['overall_rating'].sort_values(ascending=False)

# Overall rating'in kendisiyle korelasyonu 1, bu y√ºzden onu √ßƒ±karƒ±yorum
correlations = correlations.drop('overall_rating')

# En y√ºksek korelasyona sahip √∂zellikler
correlations
```

```python
# En y√ºksek korelasyona sahip ilk 10 √∂zellik
top_10_features = correlations.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_features.values, y=top_10_features.index, hue=top_10_features, palette="viridis")
plt.title('10 Attributes Most Correlated with Overall Rating')
plt.xlabel('Correlation')
plt.ylabel('Skills')
plt.tight_layout()
plt.show()
```

### The Most Important Features For Each Position

```python
position_correlations = {}

for position in df3['positions'].unique():
    position_df = df3[df3['positions'] == position]
    
    # Sayƒ±sal s√ºtunlarƒ± se√ßiyoruz
    pos_numeric_df = position_df.select_dtypes(include=['int64', 'float64'])
    
    # Korelasyon hesaplayalƒ±m
    pos_corr = pos_numeric_df.corr()['overall_rating'].sort_values(ascending=False)
    pos_corr = pos_corr.drop('overall_rating')
    
    # Bu pozisyon i√ßin t√ºm korelasyonlarƒ± saklayalƒ±m
    position_correlations[position] = pos_corr

# En √ßok g√∂r√ºlen pozisyonlarƒ± bulalƒ±m (tercihen en pop√ºler 8-10 pozisyon)
position_counts = df3['positions'].value_counts()
top_positions = position_counts.head(10).index.tolist()

# En sƒ±k g√∂r√ºlen √∂zellikleri belirleyelim (her pozisyon i√ßin en y√ºksek korelasyonlu 10 √∂zelliƒüi alalƒ±m)
top_features = set()
for position in top_positions:
    top_features.update(position_correlations[position].head(10).index.tolist())

feature_name_map = {
    'ball_control': 'Ball Control',
    'reactions': 'Reaction',
    'composure': 'Composure',
    'dribbling': 'Dribbling',
    'positioning': 'Positioning',
    'vision': 'Vision',
    'finishing': 'Finishing',
    'shot_power': 'Shot Power',
    'long_shots': 'Long Shot',
    'penalties': 'Penalties',
    'volleys': 'Volleys',
    'crossing': 'Crossing',
    'freekick_accuracy': 'Freekick Accuracy',
    'curve': 'Curve',
    'average_passing': 'Average Passing',
    'acceleration': 'Acceleration',
    'sprint_speed': 'Sprint Speed',
    'agility': 'Agility',
    'balance': 'Balance',
    'stamina': 'Stamina',
    'strength': 'Strength',
    'jumping': 'Jumping',
    'heading_accuracy': 'Heading Accuracy',
    'marking': 'Marking',
    'standing_tackle': 'Standing Tackle',
    'sliding_tackle': 'Sliding Tackle',
    'interceptions': 'Interceptions',
    'aggression': 'Aggression',
    'skill_moves(1-5)': 'Skill Moves (1-5)',
    'weak_foot(1-5)': 'Weak Foot (1-5)'
}

# En √∂nemli 15 √∂zellik
top_features_list = sorted(list(top_features), 
                           key=lambda x: sum([position_correlations[pos][x] if x in position_correlations[pos] else 0 for pos in top_positions]), 
                           reverse=True)[:15]

# Pozisyonlar ve √∂zellikler i√ßin korelasyon matrisi 
corr_matrix = np.zeros((len(top_positions), len(top_features_list)))

for i, position in enumerate(top_positions):
    for j, feature in enumerate(top_features_list):
        if feature in position_correlations[position]:
            corr_matrix[i, j] = position_correlations[position][feature]

plt.figure(figsize=(16, 12))
ax = sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='YlGnBu',
    xticklabels=[feature_name_map.get(f, f) for f in top_features_list],
    yticklabels=top_positions,
    linewidths=.5,
    fmt='.2f',
    vmin=0.4,  # Minimum koerlasyon deƒüeri
    vmax=1.0   # Maximum korelasyon deƒüeri
)

plt.title('The Most Important Features For Each Position', fontsize=16)
plt.tight_layout()

plt.xticks(rotation=45, ha='right')
plt.savefig('pozisyon_korelasyon_analizi.png', dpi=300, bbox_inches='tight')
plt.show()
```
